```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("/salaries.csv");
```

```python
df.head()
```

```python
inputs=df.drop("salary_more_then_100k",axis="columns")
inputs
```

```python
target=df['salary_more_then_100k']
target
```

```python
le_inputs=LabelEncoder()
```

```python
inputs['le_company']=le_inputs.fit_transform(inputs['company'])
inputs['le_job']=le_inputs.fit_transform(inputs['job'])
inputs['le_degree']=le_inputs.fit_transform(inputs['degree'])
inputs
```

```python
final_inputs=inputs.drop(["company","job","degree"],axis='columns')
final_inputs
```

```python
model=DecisionTreeClassifier()
model.fit(final_inputs,target)
```

```python
model.score(final_inputs,target)
```

```python
model.predict([[1,1,0]])
```

```python
import pickle
```

```python
with open("/decision_tree_sal_predict_model",'wb') as f:
  pickle.dump(model,f)
```
