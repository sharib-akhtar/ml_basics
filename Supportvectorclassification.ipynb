```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
df=load_digits()
df
```

```python
dir(df)
```

```python
inputs=pd.DataFrame(df.data,df.target)
inputs.head()
```

```python
inputs["target"]=df.target
inputs.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs.drop("target",axis='columns'),inputs.target,test_size=0.2)
```

```python
x_train.head()
```

```python
len(x_test)
```

```python
len(x_train)
```

```python
len(inputs)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='rbf',C=3,gamma='scale')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
