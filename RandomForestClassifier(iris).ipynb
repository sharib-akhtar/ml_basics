```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
dir(data)
```

```python
input=pd.DataFrame(data['data'],columns=data.feature_names)
input.head()
```

```python
input['target_species']=data.target
input.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input.drop("target_species",axis='columns'),input.target_species,test_size=0.2)
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
len(x_test)
```

```python
model=RandomForestClassifier(n_estimators=50)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
mat=confusion_matrix(y_test,y_predicted)
mat
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(mat,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```
